Visualization of Covid DataSet

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/covid_19_india.csv")
```

```python
data.head()
```

```python
plt.title("Covid 19 Deaths in India")
plt.xlabel('State',fontsize=18)
plt.ylabel('Deaths',fontsize=18)
plt.bar('State/UnionTerritory','Deaths',data=data)
plt.xticks(rotation='vertical')
plt.show()
```

```python
x=data['State/UnionTerritory']
y=data['Deaths']
plt.title("Covid 19 deaths verses State in India")
plt.xlabel("State",fontsize=18)
plt.ylabel("Deaths",fontsize=18)
plt.plot(x,y)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.title("Covid 19 Deaths in India")
plt.xlabel('State',fontsize=18)
plt.ylabel('Cured',fontsize=18)
plt.bar('State/UnionTerritory','Cured',data=data)
plt.xticks(rotation='vertical')
plt.show()
```

```python
x=data['State/UnionTerritory']
y=data['Cured']
plt.title("Covid 19 state verses recovery  in India")
plt.xlabel("State",fontsize=18)
plt.ylabel("Cured",fontsize=18)
plt.plot(x,y)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.title("Covid 19 new cases in India")
plt.xlabel('State',fontsize=18)
plt.ylabel('Confirmed',fontsize=18)
plt.bar('State/UnionTerritory','Confirmed',data=data)
plt.xticks(rotation='vertical')
plt.show()
```

```python
x=data['State/UnionTerritory']
y=data['Confirmed']
plt.title("Covid 19  State verses new cases in India")
plt.xlabel("State",fontsize=18)
plt.ylabel("Confirmed",fontsize=18)
plt.plot(x,y)
plt.xticks(rotation='vertical')
plt.show()
```

```python
x=data['Date']
y=data['Deaths']
plt.title("Covid 19 Day verses Deaths in India")
plt.xlabel("Date",fontsize=18)
plt.ylabel("Deaths",fontsize=18)
plt.plot(x,y)
plt.show()
```

```python
plt.title("Covid 19 deaths in India")
plt.xlabel('Date',fontsize=18)
plt.ylabel('Deaths',fontsize=18)
plt.bar('Date','Deaths',data=data)
plt.show()
```

```python
x=data['Date']
y=data['Cured']
plt.title("Covid 19 Day verses Recovery in India")
plt.xlabel("Date",fontsize=18)
plt.ylabel("Cured",fontsize=18)
plt.plot(x,y)
plt.show()
```

```python
plt.title("Covid 19 Day verses recovery in India")
plt.xlabel('Date',fontsize=18)
plt.ylabel('Cured',fontsize=18)
plt.bar('Date','Cured',data=data)
plt.show()
```

```python
x=data['Date']
y=data['Confirmed']
plt.title("Covid 19 Day verses new cases in India")
plt.xlabel("Date",fontsize=18)
plt.ylabel("Confirmed",fontsize=18)
plt.plot(x,y)
plt.show()
```

```python
plt.title("Covid 19 Day verses new cases in India")
plt.xlabel('Date',fontsize=18)
plt.ylabel('Confirmed',fontsize=18)
plt.bar('Date','Confirmed',data=data)
plt.show()
```
